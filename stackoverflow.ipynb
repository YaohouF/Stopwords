```python
import pandas as pd
css_data = pd.read_csv('css_question.csv')
java_data = pd.read_csv('javascript_question.csv')
deep_data = pd.read_csv('deep_learning_question.csv')
python_data = pd.read_csv('pythonv2.csv')
```

```python
css_ques = css_data.ques
java_ques = java_data.ques
deep_ques = deep_data.ques
python_ques = python_data.ques
```

```python
len(python_ques)+len(java_ques)+len(deep_ques)+len(css_ques)
```

```python
# Processing the corpus (lower, lemmatization)
import nltk
# Lower the txt
css_ques = css_ques.str.lower().str.split()
java_ques = java_ques.str.lower().str.split()
deep_ques = deep_ques.str.lower().str.split()
python_ques = python_ques.str.lower().str.split()

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in text]

css_ques.apply(lemmatize_text)
java_ques.apply(lemmatize_text)
deep_ques.apply(lemmatize_text)
python_ques.apply(lemmatize_text)
```

combine all corpus into one column

```python
corpus = []
for i in range (len(css_ques)):
    corpus.append(css_ques[i])
for i in range (len(java_ques)):
    corpus.append(java_ques[i])
for i in range (len(deep_ques)):
    corpus.append(deep_ques[i])
for i in range (len(python_ques)):
    corpus.append(python_ques[i])
```

```python
for i in range(len(corpus)):
    corpus[i] = ' '.join(corpus[i])
```

```python
corpus
```

Naive Approach: Find the most common words in corpus

```python
questions = ''
for i in range(len(css_ques)):
    if i == 0:
        questions = questions + css_ques[i]
    else:
        questions = questions + " " + css_ques[i]

```

```python
for i in range(len(java_ques)):
    questions = questions + " " + java_ques[i]
```

```python
for i in range(len(deep_ques)):
    questions = questions + " " + deep_ques[i]
```

```python
for i in range(len(python_ques)):
    questions = questions + " " + python_ques[i]
```

```python
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer()
lower_questions = questions.lower()
lower_questions = lower_questions.split()
questions_processed = [lemmatizer.lemmatize(word) for word in lower_questions]
#questions_processed = ' '.join(questions_processed)
```

```python
from collections import Counter
Counter = Counter(questions_processed)
most_occur = Counter.most_common(200)
  
print(most_occur)
```

```python
common_words = []
for i in range(len(most_occur)):
    common_words.append(most_occur[i][0])
```

```python
textfile = open("naive_approach.txt", "w")
for element in common_words:
    textfile.write(element + "\n")
textfile.close()
```

```python
# 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(corpus)
```

```python
# create dictionary to find a tfidf word each word
word_lst = vectorizer.get_feature_names()
count_lst = x.toarray().sum(axis=0)

vocab_df = pd.DataFrame((zip(word_lst,count_lst)),
                          columns= ["vocab","tfidf_value"])

sorted_df = vocab_df.sort_values(by="tfidf_value",ascending=False)
```

```python
textfile = open("tfidf_approach.txt", "w")
for element in sorted_df.vocab.head(200):
    textfile.write(element + "\n")
textfile.close()
```
